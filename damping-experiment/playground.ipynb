```python
from scipy import signal
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from scipy.fftpack import fft

t_end = 2
fs = 128
t = np.linspace(0,t_end,t_end*fs)
y = np.sin(10*2*np.pi*t)
plt.plot(t,y)
plt.show()

f = np.fft.fft(y)
freq = np.fft.fftfreq(t.shape[-1],d=1/fs)
plt.plot(freq,np.abs(f))
plt.show()

plt.plot(np.linspace(0,64,t_end*64),np.abs(f)[:t_end*64])
plt.show()
```
